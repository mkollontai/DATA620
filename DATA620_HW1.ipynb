```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
 G = nx.Graph()
```

```python
G.add_edges_from([("A","B"),("A","C"),("A","D"),("A","F"),("B","D"),("B","E"),("B","G"),("C","D"),("C","F"),("D","E"),("D","F"),("D","G"),("E","G"),("F","G"),("F","H"),("G","H"),("H","I"),("I","J")])
```

```python
nx.draw(G, with_labels = True, font_weight='bold', node_size = 500)
```
